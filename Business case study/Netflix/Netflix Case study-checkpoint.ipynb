```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the top 10 countries for Movies and TV Shows based on title counts
top_movie_countries = netflix[netflix['type'] == 'Movie']['country'].value_counts().head(10)
top_tv_countries = netflix[netflix['type'] == 'TV Show']['country'].value_counts().head(10)

plt.figure(figsize=(12, 5))
plt.suptitle("Top 10 Countries with Highest Movie / TV Show Count", fontsize=14)

# Left subplot: Movies
plt.subplot(1, 2, 1)
sns.barplot(hue=top_movie_countries.index, y=top_movie_countries.values, palette="viridis")
plt.xticks(rotation=90, fontsize=10)
plt.title("Top 10 Countries with Highest Movie Count")
plt.ylabel("Count")  # Y-axis shows the number of Movies

# Right subplot: TV Shows
plt.subplot(1, 2, 2)
sns.barplot(hue=top_tv_countries.index, y=top_tv_countries.values, palette="magma")
plt.xticks(rotation=90, fontsize=10)
plt.title("Top 10 Countries with Highest TV Show Count")
plt.ylabel("Count")  # Y-axis shows the number of TV Shows

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Group by country and type, then count the number of titles
country_counts = netflix.groupby(['country', 'type'])['title'].count().reset_index()

# Filter top 15 countries with the highest count
top_countries = country_counts.groupby('country')['title'].sum().nlargest(15).index
filtered_data = country_counts[country_counts['country'].isin(top_countries)]

# Plot
plt.figure(figsize=(12, 5))
plt.suptitle("Netflix_CaseStudy", fontsize=14, fontweight='bold')
plt.title("Country wise Movie/ TV Show Count", fontsize=12)

sns.barplot(x='country', y='title', hue='type', data=filtered_data, dodge=True, palette="deep")

plt.xticks(rotation=90, fontsize=10)
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title="Type", loc="upper right")  # Legend to differentiate Movies & TV Shows
plt.show()
```

```python
import pandas as pd

# Count movies added each year
n = l['year_added'].value_counts().reset_index()
n.columns = ['year_added', 'count']
n['count2'] = n['count'].shift(-1)  # Shift for previous year comparison
n['pct_change_movies'] = round(n['count'].pct_change(-1) * 100, 2)  # Percentage change

# Keep only the last 5 years and sort by year
p = n[:5].sort_values(by='year_added', ascending=False).drop(columns=['count', 'count2'])

# Count TV shows added each year
o = m['year_added'].value_counts().reset_index()
o.columns = ['year_added', 'count']
o['count2'] = o['count'].shift(-1)
o['pct_change_tvshows'] = round(o['count'].pct_change(-1) * 100, 2)

# Keep only the last 5 years and sort by year
q = o[:5].sort_values(by='year_added', ascending=False).drop(columns=['count', 'count2'])

# Merge both datasets on 'year_added' without count columns
final_df = p.merge(q, on='year_added', how='left')

# Display result
print(final_df)
```
