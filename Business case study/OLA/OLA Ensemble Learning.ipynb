#### DATA PROCESSING AND FEATURE ENGINEERING

```python
ola1 = ola.copy(deep=True)
```

```python
# # Target variable creation: Create a column called target which tells whether the driver has left the company- 
# # driver whose last working day is present will have the value 1

first =  (ola1.groupby('Driver_ID').agg({'LastWorkingDate':'last'})['LastWorkingDate'].isna()).reset_index()
first['LastWorkingDate'].replace({True:1,False:0},inplace=True)
first.rename(columns={'LastWorkingDate':'target'},inplace=True)
first.head()
```

```python
# Create a column which tells whether the quarterly rating has increased for that driver - 
# for those whose quarterly rating has increased we assign the value 1
QR1 =  (ola1.groupby('Driver_ID').agg({'Quarterly Rating':'first'})['Quarterly Rating']).reset_index()
QR2 = (ola1.groupby('Driver_ID').agg({'Quarterly Rating':'last'})['Quarterly Rating']).reset_index()
```

```python
QR1.shape,QR2.shape
```

```python
QR1.isna().sum(),QR2.isna().sum()
```

```python
first = first.merge(QR1,on='Driver_ID')
first = first.merge(QR2,on='Driver_ID')
```

```python
first.head()
```

```python
first['Promotion'] = np.where(first['Quarterly Rating_x'] == first['Quarterly Rating_y'], 0,1)
```

```python
# Create a column which tells whether the monthly income has increased for that driver - 
# for those whose monthly income has increased we assign the value 1
incm1 =  (ola1.groupby('Driver_ID').agg({'Income':'first'})['Income']).reset_index()
incm2 = (ola1.groupby('Driver_ID').agg({'Income':'last'})['Income']).reset_index()
```

```python
incm1.shape,incm2.shape
```

```python
incm1.isna().sum(),incm2.isna().sum()
```

```python
first = first.merge(incm1,on='Driver_ID')
first = first.merge(incm2,on='Driver_ID')
```

```python
first.head()
```

```python
first['Raise'] = np.where(first['Income_x'] == first['Income_y'], 0,1)
```

```python
first.head()
```

```python
first.tail()
```

```python
first = first[['Driver_ID','target','Raise','Promotion']]
```

```python
first.head()
```

```python
functions = {'MMM-YY':'count',
             'Driver_ID':'first',
             'Age':'max',
             'Gender':'last',
            'City':'last',
             'Education_Level':'last',
             'Dateofjoining':'first',
            'LastWorkingDate':'last',
             'Grade':'last',
             'Total Business Value':'sum',
            'Income':'sum',
             'Dateofjoining':'first',
             'LastWorkingDate':'last',
            'Joining Designation':'last',
             'Grade':'last',
             'Quarterly Rating':'first'}
ola1 = ola1.groupby([ola1['Driver_ID']]).aggregate(functions)
ola1['month'] = pd.to_datetime(ola['Dateofjoining']).dt.month
ola1['year'] = pd.DatetimeIndex(ola1['Dateofjoining']).year
ola1.rename(columns={'MMM-YY':'Reportings'},inplace=True)
```

```python
ola1.reset_index(drop=True, inplace=True)
ola1 = ola1.merge(first,on='Driver_ID')
ola1.head()
```

## ENSEMBLE LEARNING

#### Data Prepration:

The Trade-Off In general while choosing a model, we might choose to look at precision and recall scores and choose while keeping the follwing trade-off on mind : -

- If we prioritize precision, we are going to reduce our false positives. This may be useful if our targeted retention strategies prove to be expensive. We don’t want to spend unnecessarily on somebody who is not even going to leave in the first place. Also, it might lead to uncomfortable situation for the employee themselves if they are put in a situation where it is assumed that they are going to be let go/ going to leave. 

- If we prioritize recall, we are going to reduce our false negatives. This is useful since usually the cost of hiring a new person is higher than retaining n experienced person. So, by reducing false negatives, we would be able to better identify those who are actually going to leave and try to retain them by appropriate measures (competitve remuneration, engagement program, etc).

```python
from sklearn.model_selection import learning_curve
def plot_learning_curve(estimator, X, Y, title):
    train_sizes, train_scores, test_scores, _, _ = learning_curve(estimator,X,Y,return_times=True)
    fig, axes = plt.subplots(1, 1, figsize = (15, 5))
    axes.set_title(title)
    axes.plot
    axes.set_xlabel("Training examples")
    axes.set_ylabel("Score")
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    # Plot learning curve
#     32
    axes.grid()
    axes.fill_between(
    train_sizes,
    train_scores_mean - train_scores_std,
    train_scores_mean + train_scores_std,
    alpha=0.1,
    color="r",
    )
    axes.fill_between(
    train_sizes,
    test_scores_mean - test_scores_std,
    test_scores_mean + test_scores_std,
    alpha=0.1,
    color="g",
    )
    axes.plot(
    train_sizes, train_scores_mean, "o-", color="r", label="Training score"
    )
    axes.plot(
    train_sizes, test_scores_mean, "o-", color="g", label="Cross-validation score"
    )
    axes.legend(loc="best")
    plt.show()
```

## Insights:

<B>1. Workforce Demographics & Behavior</B>

- Majority of Ola drivers are male (≈58%), while females form around 42%.

- Drivers typically fall in the age group of 31–41, which is also the most active and productive segment.

- Income and business value both rise with age, but peak around 45–50 and decline afterwards, likely due to physical work fatigue.

<B>2. Education, Grade & Designation</B>

- Education Levels 1 and 2 have almost equal representation, meaning education isn't a strong differentiator in hiring.

- Most drivers join at Designation 1 (low level) → 43%, and fewer join at higher positions.

- Current grade distribution shows Grade 2 has the largest chunk (~35%), followed by Grade 1 and 3.

<B>3. Promotions, Raises & Performance</B>

- Only ~2.6% drivers received a salary raise, which is extremely low.

- ~55% received at least one promotion.

- Quarterly Ratings show most drivers get a rating of 1, which might indicate:

- Strict rating criteria

- Performance issues

- Or rating inflation/deflation problems

<B>4. Business Value & Income</B>

- Total Business Value shows a heavy positive skew with extreme outliers.

- Removing negative values (>700 drivers) improved correlations & model performance.

- Higher grades, better ratings, and more experience correspond to higher income.

<B>5. Seasonal & Temporal Trends</B>

- July has the highest driver onboarding.

- Very low onboarding during February and March.

- Significant jump in hiring post-2017 → rapid business expansion.

## Recommendations:

<B>1. Strengthen Driver Retention Through Performance-Linked Interventions</B>

- Implement targeted retention programs for drivers showing low business value, low reporting frequency, or declining ratings.

- Introduce structured performance coaching for at-risk drivers highlighted by the churn model.

- Create an early-warning dashboard based on predictive model outputs.

<B>2. Revise Compensation Structure to Improve Retention</B>

- Establish a minimum annual increment policy, as only 2–3% of drivers currently receive a raise.

- Introduce milestone-based incentives tied to business value, customer satisfaction, and reporting consistency.

- Offer higher incentives during peak months to match demand.

<B>3. Redesign Promotion Framework to Increase Motivation</B>

- Simplify the promotion pathway and communicate clear, objective criteria tied to performance metrics.

- Ensure promotion decisions are data-backed and occur at regular intervals.

- Offer small intermediate-level designations to prevent stagnation for long-tenure drivers.

<B>4. Improve Support Programs for Older Drivers</B>

- Provide age-friendly work options, such as reduced shift lengths, flexible timings, or lighter routes.

- Launch annual health screening and wellness initiatives for drivers over 45.

- Implement productivity maintenance programs to counter declining business value in higher age brackets.

<B>5. Reevaluate Quarterly Rating Mechanism</B>

- Conduct an audit of the rating system, as >50% drivers consistently receive Rating 1.

- Retrain city managers/supervisors to standardize rating distribution and eliminate rating bias.

- Align ratings with measurable KPIs to ensure fairness and consistency.

<B>6. Optimize Hiring Strategy Based on Seasonal Trends</B>

- Increase recruitment and training capacity in June–August, the period with highest driver onboarding.

- Deploy additional hiring campaigns during low-join months (Feb–Mar) to maintain operational balance.

- Align regional supply planning with city-wise demand trends.

<B>7. Strengthen Driver Engagement in High-Activity Cities</B>

- Prioritize engagement and reward programs in top driver cities (e.g., C20), which significantly contribute to total business value.

- Invest in localized support teams to reduce attrition in high-density clusters.

<B>8. Enhance Data Collection & Monitoring Infrastructure</B>

- Integrate customer feedback, cancelled rides, and route history into the driver performance dataset.

- Use telematics data to track driving behavior and personalize training programs.

- Implement unified dashboards for real-time attrition risk monitoring.

<B>9. Implement Model-Driven Retention Actions</B>

- Deploy the Gradient Boosting churn model (highest F1-score ~91%) in production for monthly churn forecasting.

- Prioritize high-recall strategies to reduce false negatives—ensuring drivers likely to leave are never missed.

- Conduct quarterly retraining of the model with updated data.

<B>10. Strengthen the Driver Relationship Lifecycle</B>

- Offer onboarding incentives for new drivers during early months when churn risk is high.

- Launch loyalty rewards for long-tenure drivers completing 2, 3, and 5-year milestones.

- Provide periodic skill-development workshops to increase driver confidence and retention.
